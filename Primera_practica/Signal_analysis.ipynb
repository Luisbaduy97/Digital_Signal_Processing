```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib qt
```

```python
data = pd.read_excel('ECG_data.xlsx')
data.head()
```

```python
plt.figure(figsize=[20,10])
plt.plot(data.Original_sig, color = 'red')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('Original ECG')
plt.show()
```

```python
plt.figure(figsize=[20,10])
plt.magnitude_spectrum(data.Original_sig, Fs=250, color = 'red')
plt.title('FFT señal original')
```

```python
plt.figure(figsize=[20,20])
plt.specgram(data.Original_sig, Fs=250)
plt.ylabel('Frequency')
plt.xlabel('Time')
```

```python
plt.figure(figsize=[20,10])
plt.plot(data.HP_sig, color = 'red')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('High Pass')
```

```python
plt.figure(figsize=[20,20])
plt.specgram(data.HP_sig, Fs=250)
plt.ylabel('Frequency')
plt.xlabel('Time')
```
